## Outlier Treatment

```python
# Select numeric columns for boxplot
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Set the style of seaborn
sns.set(style="whitegrid")

# Calculate the number of rows and columns for subplots
num_plots = len(numeric_cols)
rows = (num_plots + 3) // 4  # 4 boxplots per row

# Create a figure with subplots
plt.figure(figsize=(20, rows * 5))  # Adjust the figure size dynamically (wider for 4 plots per row)

# Add a main title
plt.suptitle("Boxplot Before Outlier Treatment", fontsize=20, y=1.02)

# Create boxplots for each numeric column
for i, col in enumerate(numeric_cols):
    plt.subplot(rows, 4, i + 1)  # Arrange plots with 4 columns per row
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot for {col}')
    plt.ylabel(col)

plt.tight_layout()
plt.subplots_adjust(top=0.9)  # Adjust layout to fit the main title
plt.show()
```

```python
# Apply IQR
columns_with_outliers = ['DataValue','LowConfidenceLimit', 'HighConfidenceLimit', 'longitude', 'latitude']

for col in columns_with_outliers:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 2 * IQR
    upper_bound = Q3 + 2 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
# Select numeric columns for boxplot
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Set the style of seaborn
sns.set(style="whitegrid")

# Calculate the number of rows and columns for subplots
num_plots = len(numeric_cols)
rows = (num_plots + 3) // 4  # 4 boxplots per row

# Create a figure with subplots
plt.figure(figsize=(20, rows * 5))  # Adjust the figure size dynamically (wider for 4 plots per row)

# Add a main title
plt.suptitle("Boxplot After Outlier Treatment", fontsize=20, y=1.02)

# Create boxplots for each numeric column
for i, col in enumerate(numeric_cols):
    plt.subplot(rows, 4, i + 1)  # Arrange plots with 4 columns per row
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot for {col}')
    plt.ylabel(col)

plt.tight_layout()
plt.subplots_adjust(top=0.9)  # Adjust layout to fit the main title
plt.show()
```

```python
X = df.drop("DataValue", axis=1)
y = df['DataValue']
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform PCA
pca = PCA(n_components=2)  # Specify the number of components
pca_result = pca.fit_transform(X_scaled)

# Create a DataFrame for PCA results
X_pca = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])

# Display PCA results
X_pca
```

##Train-Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

## Model Building

###Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Train a regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predictions
y_pred = regressor.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the decision tree classifier
model = DecisionTreeRegressor()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)


# Calculate regression metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
